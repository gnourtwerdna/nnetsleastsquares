```python
import numpy as np
import matplotlib.pyplot as plt
import util
np.random.seed(1)
```

### Part A

The stopping criterion I picked for my implenetation of the Levenberg-Marquardt algorithm is $||w_{t+1}-w_t||_2 <10^{-10}$.

```python
X_train, y_train = util.get_train_data()
w = np.random.rand(16, 1)
```

```python
losses, weights = util.LM(X_train, y_train, w, lambda_=1e-4, gamma=1e-2, activation_function='tanh')
util.loss_plot(losses, 1e-2)
```

```python
losses, weights = util.LM(X_train, y_train, w, lambda_=1e-4, gamma=1e-3, activation_function='tanh')
util.loss_plot(losses, 1e-3)
```

```python
losses, weights = util.LM(X_train, y_train, w, lambda_=1e-4, gamma=1e-4, activation_function='tanh')
util.loss_plot(losses, 1e-4)
```

I noticed that by decreasing the trust parameter in the Levenberg-Marquardt algorithm, the initial loss increased. However, each model seemed to converge to about the same loss value.

### Part B

```python
X_test, y_test = util.get_train_data(train=False)
y_predicted = util.func(X_test, weights[-1], activation_function='tanh')
test_error = util.RMSE(y_predicted, y_test)
print('Test error:{}'.format(test_error))
```
