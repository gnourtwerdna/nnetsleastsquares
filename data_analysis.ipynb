### Part A and B

The stopping criterion I picked for my implenetation of the Levenberg-Marquardt algorithm is $||w_{t+1}-w_t||_2 <10^{-10}$. The weights were initialized as random numbers from 0 to 1.

```python
X_train, y_train = util.get_train_data()
w = np.random.rand(16, 1)
```

```python
losses, weights = util.LM(X_train, y_train, w, lambda_=1e-4, gamma=1e-4, activation_function='tanh')
util.loss_plot(losses)
```

```python
y_predicted = util.func(X_train, weights[-1], activation_function='tanh')
train_error = util.RMSE(y_predicted, y_train)
print('Train error : {}'.format(train_error))

X_test, y_test = util.get_train_data(train=False)
y_predicted = util.func(X_test, weights[-1], activation_function='tanh')
test_error = util.RMSE(y_predicted, y_test)
print('Test error for Γ=1 : {}'.format(test_error))

X_test, y_test = util.get_train_data(train=False)
X_test *= 2
y_test *= 2
y_predicted = util.func(X_test, weights[-1], activation_function='tanh')
test_error = util.RMSE(y_predicted, y_test)
print('Test error for Γ=2 : {}'.format(test_error))

X_test, y_test = util.get_train_data(train=False)
X_test *= 0.5
y_test *= 0.5
y_predicted = util.func(X_test, weights[-1], activation_function='tanh')
test_error = util.RMSE(y_predicted, y_test)
print('Test error for Γ=0.5 : {}'.format(test_error))
```

The different weight initialization I tried was random numbers from -1 to 1. The model with random weights from -1 to 1 had a lot of spikes in loss throughout the early iterations. I believe the weights initialized as random numbers from 0 to 1 performed the best because it had a lower initial loss and smooth curve. Overall, a lambda value of 1e-5 with random weights from 0 to 1 performed the best.

I also tried different changing the gamma values for the test dataset to be 1, 2, and 0.5. We can see that a gamma of 0.5 gave the best test error.
